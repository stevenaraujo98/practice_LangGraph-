```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_deepseek import ChatDeepSeek
```

```python
gemini_llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
response = gemini_llm.invoke("Hola como estas?")
response.pretty_print() # response.text()
```

```python
gemini_lite_llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", temperature=0)
response = gemini_lite_llm.invoke("Hola como estas?")
response.pretty_print() # response.text()
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in spanish")
msg_2 = HumanMessage(content="Me llamo Juan")
msg_3 = AIMessage(content="Hola Juan, como estas?")
msg_4 = HumanMessage(content="Como me llamo?")
history = [msg_1, msg_2, msg_3, msg_4]
response = gemini_lite_llm.invoke(history)
response.pretty_print()
```

```python
deepseek_llm = ChatDeepSeek(
    model="deepseek-chat",
    temperature=0,
    max_tokens=500,
    timeout=None,
    max_retries=2,
)
response = deepseek_llm.invoke(history)
response.pretty_print()
```

```python
deepseek_llm = ChatDeepSeek(
    model="deepseek-reasoner",
    temperature=0,
    max_tokens=500,
    timeout=None,
    max_retries=2,
)
response = deepseek_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("google_genai:gemini-2.5-flash", temperature=0)
response = llm.invoke(history)
response.pretty_print()
```
